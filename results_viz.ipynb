```python
import csv

data = [x['total_loss'] for x in metrics_resnext_pointrend if 'total_loss' in x]

filename = 'output.csv'

with open(filename, mode='w', newline='') as file:
    fieldnames = ['resnet_pointrend']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()

    for value in data:
        writer.writerow({'resnet_pointrend': value})
```

```python
import pandas as pd

# Extract data
iter = [x['iteration'] for x in metrics_resnet if 'total_loss' in x]
rs = [x['total_loss'] for x in metrics_resnet if 'total_loss' in x]
rsx = [x['total_loss'] for x in metrics_resnext if 'total_loss' in x]
rs_pr = [x['total_loss'] for x in metrics_resnet_pointrend if 'total_loss' in x]
rsx_pr = [x['total_loss'] for x in metrics_resnext_pointrend if 'total_loss' in x]

# Find the maximum length among all lists
max_len = max(len(iter), len(rs), len(rsx), len(rs_pr), len(rsx_pr))

# Pad shorter lists with `NaN` (or `None`)
iter += [None] * (max_len - len(iter))
rs += [None] * (max_len - len(rs))
rsx += [None] * (max_len - len(rsx))
rs_pr += [None] * (max_len - len(rs_pr))
rsx_pr += [None] * (max_len - len(rsx_pr))

# Create DataFrame
df = pd.DataFrame({
    'iteration': iter,
    'resnet': rs,
    'resnext': rsx,
    'resnet_pointrend': rs_pr,
    'resnext_pointrend': rsx_pr,
})

# Save to CSV
df.to_csv('train_loss.csv', index=False)
```
