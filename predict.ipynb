```python
folder_name = "Test_Area"
site_path = "/content/drive/MyDrive/Research/data/" + folder_name
```

```python
testplot_dict = {
    # Name : [Folder, Img]
    "T0" : ["Test_Area", "test_plot_0"],
    "T1" : ["Test_Area", "test_plot_1"],
    "T2" : ["Test_Area", "test_plot_2"],
    "T3" : ["Test_Area", "test_plot_3"],
    "T4" : ["Test_Area", "test_plot_4"],
    "T5" : ["Test_Area", "test_plot_5"]
}
```

### Tile data

```python
buffer = 20
tile_width = 30
tile_height = 30
pred_dict = {}

for pa in testplot_dict:
  dataset_list = testplot_dict[pa]
  img_path = site_path + "/rgb/" + dataset_list[1] + ".tif"
  tiles_path = site_path + "/tilespred_" + pa + "/"
  # tile_data(img_path, tiles_path, buffer, tile_width, tile_height, dtype_bool = True)
  pred_dict[pa] = tiles_path
```

```python
!wget https://zenodo.org/records/10522461/files/230103_randresize_full.pth
trained_model = "./230103_randresize_full.pth"
```

```python
print(pred_dict["T5"])
```

```python
for pa in testplot_dict:
  cfg = setup_cfg(update_model=trained_model)
  predict_on_data(pred_dict[pa], predictor=DefaultPredictor(cfg))
  project_to_geojson(tiles_path, tiles_path + "predictions/", tiles_path + "predictions_geo/")
  crowns = stitch_crowns(tiles_path + "predictions_geo/", 1)
  clean = clean_crowns(crowns, 0.6, confidence=0)
  clean = clean[clean["Confidence_score"] > 0.5]
  clean = clean.set_geometry(clean.simplify(0.3))
  clean.to_file(site_path + "/" + "crown_pred_" + pa + ".gpkg")
```

```python
cfg = setup_cfg(update_model=trained_model)
predict_on_data(pred_dict["T5"], predictor=DefaultPredictor(cfg))
project_to_geojson(tiles_path, tiles_path + "predictions/", tiles_path + "predictions_geo/")
crowns = stitch_crowns(tiles_path + "predictions_geo/", 1)
clean = clean_crowns(crowns, 0.6, confidence=0)
clean = clean[clean["Confidence_score"] > 0.2]
clean = clean.set_geometry(clean.simplify(0.3))
clean.to_file(site_path + "/" + "crown_pred_" + "T502" + ".gpkg")
```
