```python
def crown_area(crown_path, out_crown):

    crown_file = gpd.read_file(crown_path)
    crown_list = list(crown_file.geometry)

    crown_copy = crown_file.copy()
    crown_copy["crown_area"] = 0.00


    for p_idx, crown_poly in enumerate(crown_list):
        crown_copy.loc[p_idx, "crown_area"] = crown_poly.area

    crown_copy.to_file(out_crown)
```

```python
def crown_area_plot(crown_path, color, plot_title):

    crown_file = gpd.read_file(crown_path)

    hist_plot = crown_file.plot.hist(
        column=["crown_area"],
        legend=False,
        stacked=False,
        alpha=0.7,
        histtype='bar',
        bins=50,
        edgecolor='black',
        linewidth=1,
        color=color,
        figsize=(6, 5)
    )


    hist_plot.set_title("Crown Area Distribution - " + plot_title, fontsize=11, pad=20)
    hist_plot.set_xlabel("Crown Area (m²)", fontsize=10)
    hist_plot.set_ylabel("Number of Trees", fontsize=10)
    hist_plot.set_xlim(left=0)
    # hist_plot.set_xlim(0, 1400)

    x_min, x_max = hist_plot.get_xlim()
    hist_plot.set_xticks(np.arange(0, x_max, 100))
    hist_plot.tick_params(axis='both', which='major', labelsize=10)

    hist_plot.grid(axis='y', alpha=0.4)
    hist_plot.legend(["Crown Area"], fontsize=9)

    plt.tight_layout()

    plt.show()
```

```python
def crown_area_plot2(crown_path, color, plot_title, code_name):

    crown_file = gpd.read_file(crown_path)

    hist_plot = crown_file.plot.hist(
        column=["crown_area"],
        legend=False,
        stacked=False,
        alpha=0.7,
        histtype='bar',
        bins=80,
        edgecolor='black',
        linewidth=1,
        color=color,
        figsize=(6, 5)
    )

    code_title = f"({code_name})"

    hist_plot.set_title(plot_title + " " + code_title, fontsize=11, pad=10)
    hist_plot.set_xlabel("Luas Tajuk (m²)", fontsize=10)
    hist_plot.set_ylabel("Jumlah Pohon", fontsize=10)
    hist_plot.set_xlim(left=0)
    hist_plot.set_xlim(0, 1400)

    x_min, x_max = hist_plot.get_xlim()
    hist_plot.set_xticks(np.arange(0, x_max, 100))
    hist_plot.tick_params(axis='both', which='major', labelsize=10)

    hist_plot.grid(axis='y', alpha=0.4)
    # hist_plot.legend(["Crown Area"], fontsize=9)

    plt.tight_layout()

    plt.show()
```

```python
def crown_area_plot3(crown_path, color, plot_title, code_name, bin_num):

    crown_file = gpd.read_file(crown_path)

    hist_plot = crown_file.plot.hist(
        column=["crown_area"],
        legend=False,
        stacked=False,
        alpha=0.7,
        histtype='bar',
        bins=bin_num,
        edgecolor='black',
        linewidth=1,
        color=color,
        figsize=(6, 5)
    )

    code_title = f"({code_name})"

    hist_plot.set_title(plot_title + " " + code_title, fontsize=11, pad=10)
    hist_plot.set_xlabel("Luas Tajuk (m²)", fontsize=10)
    hist_plot.set_ylabel("Jumlah Pohon", fontsize=10)
    hist_plot.set_xlim(left=0)
    hist_plot.set_xlim(0, 1400)

    x_min, x_max = hist_plot.get_xlim()
    hist_plot.set_xticks(np.arange(0, x_max, 100))
    hist_plot.tick_params(axis='both', which='major', labelsize=10)

    hist_plot.grid(axis='y', alpha=0.4)
    # hist_plot.legend(["Crown Area"], fontsize=9)

    plt.tight_layout()

    plt.show()
```
